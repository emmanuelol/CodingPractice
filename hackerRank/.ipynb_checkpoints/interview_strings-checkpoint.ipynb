```python
def mapas(s):
    mapa={}
    for i in s:
        if i not in mapa:
            mapa[i]=1
        else:
            mapa[i]+=1
    return mapa
```

```python
def isValid(s):
    mapa=mapas(s)
    temp=list(mapa.values())
    if min(temp)!=max(temp):
        maximo=max(temp)-1
        if min(temp)==maximo:
            result='YES'
        else:
            result='NO'
    return result
```

```python
def commonChild(s1, s2):
    n1=len(s1)
    n2=len(s2)

    lcs=[[0]*(n2+1) for i in range(n1+1)]
    for i in range(n1):
        for j in range(n2):
            if s1[i]==s2[j]:
                lcs[i][j]=lcs[i-1][j-1] +1
            else:
                lcs[i][j]= max(lcs[i-1][j], lcs[i][j-1])
    return lcs[n1-1][n2-1]
```

```python
commonChild(string_1_a,string_1_b)
```

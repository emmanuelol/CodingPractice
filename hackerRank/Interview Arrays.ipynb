```python

#array=[2,3,4,1,5]
# output 3
###
array=[2,31,1,38,29,5,44,6,12,18,39,9,48,49,13,11,7,27,14,33,50,21,46,23,15,26,8,47,40,3,32,22,34,42,16,41,24,10,4,28,36,30,37,35,20,17,45,43,25,19]
#output=46
def minimum_swap(array):
    if array is None:
        raise ValueError('required at least two element')
    count=0
    for i in range(1,len(array)+1):
        
        if array[i-1]!=i:
            print('i :',i)
            print('index :',i-1)
            print('array[i]:',array[i-1])
            #temp=i
            #array[i-1],temp=temp,array[i-1]
            temp=array.index(i)
            print('array index :',array[temp])
            array[i-1],array[temp]=array[temp],array[i-1]
            print('new array[i]',array[i-1])
            count+=1
    return count
```

* This solution create a hash with the values of the array as keys, the index as the value
* the swaping is using the current position of the array and the hash to do a version of swaping in place
#### complexity
* O(n+n) time
* O(n) space

```python
def minimumSwaps(arr):
    temp = [0] * (len(arr) + 1)
    for pos, val in enumerate(arr):
        temp[val] = pos
        pos += 1
    swaps = 0
    for i in range(len(arr)):
        if arr[i] != i+1:
            swaps += 1
            t = arr[i]
            arr[i] = i+1
            arr[temp[i+1]] = t
            temp[t] = temp[i+1]
    return swaps
```

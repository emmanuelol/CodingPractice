```python
def containsDuplicate(nums):
    if nums is None:
        raise valueError('add a list')
    numbers_seen=set()
    for num in nums:
        if num not in numbers_seen:
            numbers_seen.add(num)
        else:
            return True
    return False
```

```python
containsDuplicate(test1)
```

```python
containsDuplicate(test2)
```

```python
containsDuplicate(test3)
```

```python
def singleNumber(nums):
    if nums is None:
        raise valueError('add a list')
    numbers_seen=set()
    for num in nums:
        if num not in numbers_seen:
            numbers_seen.add(num)
        else:
            numbers_seen.remove(num)
    return numbers_seen.pop()
```

```python
nums1 = [1,2,2,1]
nums2 = [2,2]
```

```python
def intersectionV2(nums1,nums2):
    set1=set(nums1)
    set2=set(nums2)
    return list(set1&set2)
```

```python
intersectionV2(nums1,nums2)
```

```python
def intersectionV1(nums1,nums2):
    mapa={}
    for num in nums1:
        if num not in mapa:
            mapa[num]=1
        else:
            mapa[num]+=1
    result=[]
    for num in nums2:
        if num in mapa:
            result.append(num)
            mapa.pop(num)
    return result
```

```python
intersectionV1(nums1,nums2)
```

```python
def get_next(n):
    total_sum=0
    while(n):
        total_sum+=(n%10) ** 2
        n=n//10
        print(n)
        print(total_sum)
    return total_sum
```

```python
def happy_number(n):
    slow=n
    fast=get_next(n)
    while fast!=1 and slow!=fast:
        slow=get_next(n)
        fast=get_next(get_next(n))
    return fast==1
```

```python
happy_number(19)
```
